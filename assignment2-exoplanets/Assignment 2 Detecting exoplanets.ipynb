```python
# Create a new dataframe with percentage of missing values in each column

missing_values = pd.DataFrame(exoplanet_df.isnull().sum()*100/len(exoplanet_df), columns=['% of missing values'])

missing_values.sort_values(by='% of missing values', ascending=False)
```

```python
# Make bar plot of top five missing values

missing_values.sort_values(by='% of missing values', ascending=False).head(5).plot(kind='bar')
```

```python
# Calculate number of entries that lies above the third quartile and how many values lies below the first quartile
# for all numeric value columns

# put the quartiles and number of outliers in a dataframe
quartiles = pd.DataFrame(columns=['75%', '25%', 'Outliers'])

for col in exoplanet_df.select_dtypes(include='number').columns:
    quartiles.loc[col] = [exoplanet_df[col].quantile(0.75), exoplanet_df[col].quantile(0.25), len(exoplanet_df[exoplanet_df[col] > exoplanet_df[col].quantile(0.75) + 1.5*(exoplanet_df[col].quantile(0.75) - exoplanet_df[col].quantile(0.25))])]

quartiles
```
