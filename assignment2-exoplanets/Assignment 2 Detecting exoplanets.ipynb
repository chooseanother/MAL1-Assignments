# Split Data

```python
from sklearn.model_selection import train_test_split

# Divide the data into training and test
target = exoplanet_df['KeplerDispositionStatus']
KOI_data = exoplanet_df.drop(['KeplerDispositionStatus','ArchiveDispositionStatus'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(
    KOI_data, target, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Fix the parameter space
parameters = {
    'n_neighbors': range(1, 20), 
    "metric": ['euclidean','manhattan','chebyshev','minkowski',]}
grid_search = GridSearchCV(KNeighborsClassifier(), parameters, cv=5, return_train_score=True)
```

```python
grid_search.fit(X_train.values, y_train.values)
```

```python
print("Test score: {:.2f}".format(grid_search.score(X_test.values, y_test.values)))
```

```python
print("Best parameter: {}".format(grid_search.best_params_))
```

```python
print("Best cross-validation score: {}".format(grid_search.best_score_))
```

```python
print("Best estimator: {}".format(grid_search.best_estimator_))
```

# Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

def model_scores(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print("Accuracy score: {:.2f}".format(accuracy_score(y_test, y_pred)))
    print("Precision score: {:.2f}".format(precision_score(y_test, y_pred)))
    print("Recall score: {:.2f}".format(recall_score(y_test, y_pred)))
    print("F1 score: {:.2f}".format(f1_score(y_test, y_pred)))
    # calculate specificity score
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    specificity = tn / (tn+fp)
    print("Specificity score: {:.2f}".format(specificity))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

def plot_confusion_matrix(y_true, y_pred):
    """
    This function plots the confusion matrix
    """
    fig, ax = plt.subplots(figsize=(8, 8))
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='d')
    plt.ylabel('Actual label')
    plt.xlabel('Predicted label')
    plt.title('Confusion matrix')
    plt.show()
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

def plot_precision_recall_curve(y_true, y_pred):
    """
    This function plots the precision-recall curve
    """
    precision, recall, thresholds = precision_recall_curve(y_true, y_pred)
    plt.plot(precision, recall)
    plt.xlabel('Precision')
    plt.ylabel('Recall')
    plt.title('Precision-Recall Curve')
    plt.show()
```

```python
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

def plot_roc_curve(y_true, y_pred):
    """
    This function plots the ROC-curve
    """
    fpr, tpr, thresholds = roc_curve(y_true, y_pred)
    plt.plot(fpr, tpr)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.show()
```

```python
model_scores(grid_search.best_estimator_, X_train.values, X_test.values, y_train.values, y_test.values)
```

```python
plot_confusion_matrix(y_test.values, grid_search.predict(X_test.values))
```

```python
plot_precision_recall_curve(y_test.values, grid_search.predict(X_test.values))
```

```python
plot_roc_curve(y_test.values, grid_search.predict(X_test.values))
```
